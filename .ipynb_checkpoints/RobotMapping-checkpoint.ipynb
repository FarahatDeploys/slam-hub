In this lecture, we start by defining the word robot mapping by dividing them into their parts


- Robot: A device or system that moves through the enviroment.
- Mapping: Representation or modelling of the enviroment.

For the robot to navigate through the enviroment, a series of operations are carried out 

A state estimation for the map and location (localization and mapping) followed by a navigation and motion planning therefore we need to define set of terms


State estimation: is the process of determining the internal state of a system when you cannot measure it directly.
Example of this (position x,y,z  orientation yaw,bitch,roll   velocity   acceleration   location of landmarks)


Localization: is the process of determining a robotâ€™s position and orientation within a known map.
In other words, the localization is a state estimation process for the location of the robot within a known enviroment
