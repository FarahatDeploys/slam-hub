# Quick Recap

- We are trying to estimate the robot current pose $(x_{t})$ while making use of all data we have $z_{1:t}$ and $u_{1:t}$
- Bayes filter is a framework that allows this through a recursive operation from time t up to time t-1
- This general recursive step can be generally divided into two steps
- Predcition step, this one relies on the previous state belief posterior and the transition probability: $$\int \underbrace{p(x_t \mid x_{t-1}, u_{t-1})}_{\text{Transition Probability}}
\;
\underbrace{\text{bel}(x_{t-1})}_{\text{Previous Belief}}
\; dx_{t-1}$$
- Measurement or correcction step, this relies on the measurement uncertanity
  $$\underbrace{p(z_t \mid x_t)}_{\text{Measurement Model}}$$

- These calculations are computationally expensive, to estimate the current state set of techniques can be used but they come at the cost of establishing certain assumption.
- Such assumption will influence the accuracy of the estimated state, always keep in mind that there is no such thing as free launch


# Kalman Filter

We are trying to estimate the current state of the robot by calculating **the prediction step** and the **correction step**.

These steps are computationally expensive, so we need some sort of simplification, which can be done by proposing certain assumptions for the probability distribution of the transition and the measurement.

After these assumptions, the complex integration can be easily mapped to a very compact form of matrix multiplication, which is suitable for a robot computations.


## Kalman Assumptions 
- the probability distrubtion for all informatin is guassian (measurement, prediction), as this will enable a guassian marginalization and conditiong
- linear motion / transition model, this is essential to preseve the initial guassian distribution from time step (1) up to (t)

- The motion model can be assumed as follows:
$$
x_{t} = A_{t}x_{t-1}+B_{t}u_{t-1}+\epsilon
$$
- Measurement model can also be representd throught the following equation
$$
z_{t} = C_{t}x_{t}+\delta_t
$$

Hint: such equation doesn't empose a certain model for the motion or measurement but only limits it to a linear general model for the measurement and the motion



- these equations are then injected into the probaility distruibtion equaiton of the normal distrubition, for example let's assume we are willing to represent the nomral distribution of variable x

 


# From the Assumption to Algorithm

We start from the generic **Bayes Filter formulation**:

## 1. Bayes Filter Equations

**Prediction**  
$ bel'(x_t)=\int p(x_t|x_{t-1},u_t)\,bel(x_{t-1})\,dx_{t-1} $

**Correction**  
$ p(x_t|z_t)=\eta\, p(z_t|x_t)\,bel'(x_t) $

---

## 2. Kalman Filter Assumptions (Rules)

To make the Bayes filter solvable analytically, we use **Gaussian + linear model assumptions**:

### **Rule 1 — State is Gaussian**

$ p(x_{t-1}) =\mathcal{N}(\mu_{t-1},\,\Sigma_{t-1}) $

### **Rule 2 — Motion model for transition is linear with Gaussian noise**

$ x_t = A x_{t-1} + B u_t + w_t,\qquad w_t \sim \mathcal{N}(0,Q) $

Probability transition form:  
$ p(x_t|x_{t-1},u_t)=\mathcal{N}(A x_{t-1}+B u_t,\; Q) $

### **Rule 3 — Measurement model is linear with Gaussian noise**

$ z_t = H x_t + v_t,\qquad v_t \sim \mathcal{N}(0,R) $

Probability form:  
$ p(z_t|x_t)=\mathcal{N}(H x_t,\; R) $

---

## 3. Substituting Gaussians into Bayes Filter

### **Prediction Step**

$ p(x_t)=\int \mathcal{N}(A x_{t-1}+B u_t,Q) \cdot \mathcal{N}(\mu_{t-1},\Sigma_{t-1})\,dx_{t-1} $

> Product of Gaussians + integration → Gaussian

Kalman Prediction Results:  
$ \mu_t^- =A\mu_{t-1}+Bu_t $  
$ \Sigma_t^- =A\Sigma_{t-1}A^T+Q $

---

### **Update Step**

$ p(x_t|z_t)=\eta \cdot \mathcal{N}(H x_t,R) \cdot \mathcal{N}(\mu_t^-,\Sigma_t^-) $

> Product of Gaussians → Gaussian

Kalman Update Equations:  
$ K_t=\Sigma_t^-H^T(H\Sigma_t^-H^T+R)^{-1} $  
$ \mu_t=\mu_t^-+K_t(z_t-H\mu_t^-) $  
$ \Sigma_t=(I-K_tH)\Sigma_t^- $

## limitation of the kalman filter 
- non linear observation function, that maps from the observation to the state in a non linear manner resulting non guassian result 
- non linear motion transition model, which will result a non guassian distrubtion as result of multiplication
- This can be formulated mathmatically by:
  $x_{t} = f(x_{t-1},u_{t})+ \epsilon $ and the measurement to be $z = g(x_{t})+ \epsilon$ , where f and g are non linear functions
